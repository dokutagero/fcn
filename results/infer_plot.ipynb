```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def report(csv_file):
    df = pd.read_csv(csv_file)
    delamination_df = df['delamination'] * 100
    rebar_df = df['rebar_exposure'] * 100
    print('Delamination Mean IoU: {}'.format(delamination_df.dropna().mean()))
    print('Rebar exposure Mean IoU: {}'.format(rebar_df.dropna().mean()))
    mean = np.mean( np.array([ delamination_df.dropna().mean(), rebar_df.dropna().mean() ]), axis=0 )
    print('Mean IoU: {}'.format(mean))
    delamination_df.dropna().plot(kind='hist', alpha=0.5, bins=40, normed=False)
    rebar_df.dropna().plot(kind='hist', alpha=0.5, bins=40, normed=False)
```

```python
plt.figure(1)
report('infer_with_aug.csv')
plt.figure(2)
report('infer_without_aug.csv')
```
